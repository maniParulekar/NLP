### 2.       Extract noun phrase (NP) chunks from your reviews using the following procedure:
a.       In Python, use BeautifulSoup to grab the main review text from each link.  
b.       Next run each review text through a tokenizer, and then try to NP-chunk it with a shallow parser. 


```python
from nltk.chunk import RegexpParser

#---The following NP chunk rule created from http://www.nltk.org/book/ch07.html
chunk_grammar = r"""
NP: {<DT>?<JJ>*<NN.*>}
    {<DT|PP\$>?<JJ>*<NN>}   # chunk determiner/possessive, adjectives and noun
    {<NNP>+} # chunk sequences of proper nouns
    {<NN.*>+}
    {<N.*>{2,}}
    {<NN.*>+<CD>}             
"""
rTextList= list(df.rText)
rTitleList= list(df.rTitle)
mNameList= list(df.MovieName)
##---------------------Custom RegEx Chunker Funtion--------------------------------------------
## The following funtion code created from http://www.nltk.org/book/ch07.html
def customRegexChunker(mNameL, rTitleL, rTextL):
    
    for (mName, rTitleStr, rStr) in zip(mNameL, rTitleL, rTextL): 
    
        print ("\n***************************************************************************")
        print ("\tMovie: ", mName, "Title: ", rTitleStr)
        print ("***************************************************************************")
    
        if (rStr == "NA"):
            rStr = rTitleStr
        else:
            rStr = rTitleStr + ". " +rStr
            
        rStr = nltk.word_tokenize(rStr)
        rStr = nltk.pos_tag(rStr)
        
        rc = RegexpParser(chunk_grammar)
        c = rc.parse(rStr)
        for subtree in c.subtrees():
            if subtree.label() == 'NP': 
                print(subtree)

    return
```
